## Introducción: *Vehicule Routing*

Sea un digrafo $G=(V,A)$ y una función de costo en sus arcos, digamos $w:A\to\mathbb{N}$. Sea $t$ un vértice de $G$.  Considere un conjunto de vértices $C$ de $G$ que no contiene a $t$ y cuyos elementos representan domicilios de personas que deben ser recogidos y trasladados a $t$ siguiendo un dicamino en $G$.  Los radio-taxis disponibles tienen una capacidad para transportar a lo más $m\in\mathbb{N}^{\star}$ pasajeros simultáneamente. Se desea transportar al destino $t$ a todas las personas minimizando el costo total. El costo por radio-taxi está dado por la distancia recorrida más un costo fijo $c_f$ (la “bajada de bandera”) .  El costo total de los traslados es la suma de los costos de todos los radio-taxis.

Para formular el problema descrito como un PE se propone lo siguiente:
1. Extender $G$ con un vértice adicional $s\not\in V$, y arcos de costo $0$ de $s$ a $v$, cualquiera sea $v\in V$. El grafo así obtenido lo seguiremos denotando por $G$, sus vértices por $V$ y sus arcos por $A$. Por ejemplo, a continuación se muestra un grafo $G$ aumentado por un vértice $s$ (los valores de $w$ para cada arco se muestran como etiquetas en las aristas): 
<img src="fig1TP.png" alt="Alternative text" />
2. Construir un nuevo digrafo $\widehat{G}=(\widehat{V},\widehat{A})$ con costos $\widehat{w}:\widehat{A}\to\mathbb{N}$ en los arcos de $\widehat{G}$ tal que:
    * el conjunto de vértices $\widehat{V}$ es $C\cup\{s,t\}$,
    * el conjunto de arcos $\widehat{A}$ es tal que $(u,v)\in\widehat{A}$ si $u=s$ y $v\in C$, o $u\in C$ y $v=t$, o $u,v\in C$ y existe un dicamino en $G$ de $u$ a $v$,
    * la función de costos $\widehat{w}$ es tal que $\widehat{w}((u,v))$ es el costo del $(u,v)$-dicamino de costo mínimo en $G$ con respecto a la función de costo $w$ en los arcos de $G$.
Por ejemplo, para el grafo $G$ ilustrado más arriba y tomando $C=\{2,3\}$ se obtiene el grafo $\widehat{G}$ y $\widehat{w}$ que se muestran a continuación:
<img src="fig2TP.png" alt="Alternative text" />
3. Definir la colección $\mathcal P$ de $(s,t)$-dicaminos en $\widehat{G}$ de largo a lo más $(m+1)$. Cada elemento de $\mathcal{P}$ determina una ruta factible en $G$ de un radio-taxi que traslada al destino $t$ a las a lo más $m$ personas representadas por los elementos en $\widehat{V}(p)\setminus\{s,t\}$. (Un dicamino $p$ de $\widehat{G}$ se entiende como un subgrafo de $\widehat{G}$, por lo que tiene sentido referirse a sus vértices  por $\widehat{V}(p)$).

4. Denotar el costo del dicamino  $p\in\mathcal{P}$ por $\widehat{w}(p):=\sum_{e\in \widehat{E}(p)} \widehat{w}(e)$.
5. Denotar por $\mathbb{I}_p\in\{0,1\}^C$ el vector indicatriz de los vértices en $C$ visitados por $p\in\mathcal{P}$.


Finalmente, formulamos el problema que nos interesa como sigue: 
$$
\begin{array}{rll}
(MP): & \min_x \sum_{p\in\mathcal{P}}(c_f+\widehat{w}(p))x_p \\
& \sum_{p\in\mathcal{P}} \mathbb{I}_p(v)x_p \geq 1, & \forall v\in C,\\
& x_p\in\mathbb{N}, & \forall p\in\mathcal{P}.
\end{array}
$$
Reemplazando las restricciones $x_p\in\mathbb{N}$ por $x_p\in\mathbb{R}_+$ y $\mathcal{P}$ por $\mathcal{Q}$ donde $\mathcal{Q}\subseteq\mathcal{P}$ se obtiene el siguiente *Restricted Master Problem*: 
$$
\begin{array}{rll}
(RMP): & \min_x \sum_{p\in\mathcal{Q}}(c_f+\widehat{w}(p))x_p \\
& \sum_{p\in\mathcal{Q}} \mathbb{I}_p(v)x_p \geq 1, & \forall v\in C, & (*)\\
& x_p\geq 0, & \forall p\in\mathcal{Q}.
\end{array}
$$

* un arreglo `Paths` que representa/codifica $\mathcal{Q}\subseteq\mathcal{P}$. Cada elemento `P` de `Paths` es un arreglo de a lo más $m+1$ elementos. Cada elemento toma valores en $[n]$. Por ejemplo, si $\mathcal{Q}$ consistiera de los dicaminos $s\to 3\to 2\to t$ y $s\to 2\to t$ del grafo $\widetilde{G}$ de la Introducción, entonces:

````julia
using DelimitedFiles, JuMP, Gurobi;           
dirInst = "dirTP";            # Directorio con el archivo de instancias para el ejercicio 1.
const GUROBI_ENV = Gurobi.Env() 

function funcRMP(W,Paths,Pcst)

    rmp = Model(optimizer_with_attributes(() -> Gurobi.Optimizer(GUROBI_ENV), "OutputFlag" => 0, "Presolve" => 0)); 

    function indicatriz(path, ver)
        if ver in path
            return 1
        else
            return 0
        end   
    end

    numberofpaths = length(Paths)
    numberofvertices = size(W)[1] - 2
    @variable(rmp, x[1:numberofpaths] ≥ 0)
    @objective(rmp, Min, sum(x[i]*Pcst[i] for i in 1:numberofpaths))
    cnstr = @constraint(rmp, [v in 1:numberofvertices], sum(x[i]*indicatriz(Paths[i],v) for i in 1:numberofpaths) ≥ 1)    

    return rmp, cnstr;  # cnstr[] debe ser un arreglo de referencias a las |C| restricciones (*) de (RMP).
                        # Ver: https://jump.dev/JuMP.jl/stable/manual/constraints/#jump_constraints
end
````

````julia

function path_cost(path, W)
    return sum(W[path[i], path[i+1]] for i in 1:length(path)-1)
end

function pathtoend(vertex, n) # Aprovechando que todo nodo de C tiene una arista hacia t (spg se puede asumir agregando una arista con el costo de un v-t dicamino en G).
    t = n
    path = [1, vertex, t]
    return path
end


tmpFrm=[]; tmpRes=[]; cardC=[]; val=[]; name=[]; sts=[];
for fn in readdir(dirInst)
    W = readdlm(joinpath(dirInst,fn), ',', Int64, '\n')
    n = size(W,1);
    s = 1;
    t = n;
    Paths = [];  Pcst = [];
    for v in 2:n-1
        path = pathtoend(v, n)
        push!(Paths, path)
        push!(Pcst, path_cost(path, W))
    end
    
    append!( name, [fn] );                                          # Guarda nombre del archivo con instancia
    append!( tmpFrm, @elapsed mdl,cnstr=funcRMP(W,Paths,Pcst) );    # Guarda tiempo requerido en formular (RMP)
    append!( tmpRes, @elapsed optimize!(mdl) );	                    # Guarda tiempo requerido en resolver (RMP)
    append!( cardC, size(W,1)-2 );                                  # Guarda cardinalidad de conjunto C
    if termination_status(mdl) == MOI.OPTIMAL                       # Guarda status de término y valor objetivo.
        append!( sts, ["optimo"] );
        append!( val, objective_value(mdl) );
    elseif termination_status(mdl) == MOI.INFEASIBLE
        append!( sts, ["infactible"] );
        append!( val, [Inf]);
    elseif termination_status(mdl) == MOI.DUAL_INFEASIBLE
        append!( sts, ["no_acotado"] );
        append!( val, [-Inf]);
    end           
end

# Imprime estadísticas relativas a las instancias resueltas.
println()
for i in 1:length(name)
  println( "Instancia = ", name[i] );
  println( "  |C| = ", cardC[i] );
  println( "  Tmp. formulación = ", tmpRes[i] );
  println( "  Tmp. resolución = ", tmpRes[i] );
  println( "  valor obj. = ", val[i] );
  println( "  status = ", sts[i] );
  println()
end
````

Respuesta Ejercicio 2a: 

Calculamos el dual de $(RMP)$:

Notamos que podemos escribir la función objetivo de $(RMP)$ como $\textrm{max } c^Tx$ con $x$ vector de componentes $x_p$ y $c$ vector de componentes $-c_f - \widehat{w}(p)$ para $p \in \mathcal{Q}$.

La restricción $(*)$ se puede escribir como $Ax \leq b$ con $A$ matriz de componentes $A_{v,p} = -\mathbb{I}_p(v)$ para $v \in C$ y $p \in \mathcal{Q}$ y $b$ vector de componentes $b_v = -1$ para $v \in C$.

Por lo tanto, el dual de $(RMP)$ es $\textrm{min } b^Ty$ con $y$ vector de componentes $y_v$ para $v \in C$, con restricciones $A^Ty \geq c$ e $y \geq 0$.



2b) Verifique que, si denotamos por $z^{(p)}=(z_e^{(p)} : e\in\widehat{A})\in\{0,1\}^{\widehat{A}}$ el vector indicatriz de los arcos de $\widehat{G}$ ocupados por el dicamino $p\in\mathcal{P}$, el *Pricing Problem* al que se llega tiene la forma:
$$
(PP):\ \eta = \min_{p\in\mathcal{P}}\big\{\sum_{e=(u,v)\in\widehat{A}} \widehat{\gamma}(e)z_e^{(p)}\big\}.
$$
Para todo $e=(u,v)\in\widehat{A}$, indique el valor de $\widehat{\gamma}(e)$ en función de $\widehat{w}$ y las variables duales $(y_v : v \in C)$.

### Respuesta Ejercicio 2b: 

En efecto, notemos primero que reemplazando las definiciones de $A^T$ y de c en la restricción del dual (y considerando $c_f = 0$) podemos escribir la restricción como $\sum_{e \in \widehat{E}(p)} w(e) - \sum_{v \in \widehat{V}} \mathbb{I}_p(v) y_v \geq 0$ para todo $p \in P$, por lo que el pricing problem es $\eta = \min_{p \in P} \sum_{e \in \widehat{E}(p)} w(e) - \sum_{v \in \widehat{V}} \mathbb{I}_p(v) y_v$.

Ahora, definiendo $\widehat{\gamma}((v,w))$ como $\widehat{\gamma}((v,w)) := w(v,w) - y_w$ para $v \neq s$ y como $\widehat{\gamma}((v, w)) := -(y_v + y_w)$ para $v=s$, se verifica que $\sum_{e \in \widehat{E}(p)} w(e) - \sum_{v \in \widehat{V}} \mathbb{I}_p(v) y_v = \sum_{e \in \widehat{E}(p)} \widehat{\gamma}(e) z_e^{(p)}$ con lo que $\eta = \min_{p \in P} \sum_{e \in \widehat{E}(p)} \widehat{\gamma}(e) z_e^{(p)}$. $\blacksquare$
